```python
def non_normal_outliers(col):
    IQR=df[col].quantile(0.75)-df[col].quantile(0.25)
    lower_bridge=df[col].quantile(0.25)-(IQR*1.5)
    upper_bridge=df[col].quantile(0.75)+(IQR*1.5)
    #print(lower_bridge), print(upper_bridge)
    df.loc[df[col]>upper_bridge,col]=upper_bridge
    if lower_bridge>0:
        df.loc[df[col]<lower_bridge,col]=lower_bridge
    #check_outliers(df[col])
def normal_outliers(col):
    upper_boundary=df[col].mean() + 3* df[col].std()
    lower_boundary=df[col].mean() - 3* df[col].std()
    #print(lower_boundary), print(upper_boundary),print(df[col].mean())
    df.loc[df[col]>upper_boundary,col]=upper_boundary
    if lower_boundary>0:
        df.loc[df[col]<lower_boundary,col]=lower_boundary
    #check_outliers(df[col])
    
```

```python
#since there are less number of 1s it is not good to replace all variables with the mean. since the data is imbalance.
#so replacing 0 to mean,median,random it with respect to their class variable (by checking their effects to data set )
df.groupby(df['Class Variable'])['Insulin'].mean()
```

```python
#we can observe from above that the median and mean values for diabetic, and non diabetic person varies vastly.so it is good 
#to replace with respect to their means,median
#we can also place the random values and look the data.
#treating insulin value
def check_dist(x,y):
    fig = plt.figure()
    ax = fig.add_subplot(111)
    df[x].plot(kind='kde', ax=ax)
    df[y].plot(kind='kde', ax=ax, color='red')
    lines, labels = ax.get_legend_handles_labels()
    ax.legend(lines, labels, loc='best')
```

```python
#equating insulin with mean and median values with respect to thier class variables
#we can take values as overall mean, or median but here variance in values is more so taking it individually
df['Plasma glucose'+"_mean"]=df['Plasma glucose'].copy()
df.loc[(df['Class Variable'] == 0 ) & (df['Plasma glucose'+"_mean"]==0), 'Plasma glucose'+"_mean"] = 110.283071
df.loc[(df['Class Variable'] == 1 ) & (df['Plasma glucose'+"_mean"]==0), 'Plasma glucose'+"_mean"] = 140.669687
check_dist('Plasma glucose_mean','Plasma glucose')
#median
df['Plasma glucose'+"_median"]=df['Plasma glucose'].copy()
df.loc[(df['Class Variable'] == 0 ) & (df['Plasma glucose'+"_median"]==0), 'Plasma glucose'+"_median"] = 107
df.loc[(df['Class Variable'] == 1 ) & (df['Plasma glucose'+"_median"]==0), 'Plasma glucose'+"_median"] = 140
check_dist('Plasma glucose_median','Plasma glucose')
```

```python
#equating insulin with mean and median values with respect to thier class variables
df['Insulin'+"_mean"]=df['Insulin'].copy()
df.loc[(df['Class Variable'] == 0 ) & (df['Insulin'+"_mean"]==0), 'Insulin'+"_mean"] = 68.792000
df.loc[(df['Class Variable'] == 1 ) & (df['Insulin'+"_mean"]==0), 'Insulin'+"_mean"] = 100.335821
check_dist('Insulin_mean','Insulin')
#median
df['Insulin'+"_median"]=df['Insulin'].copy()
df.loc[(df['Class Variable'] == 0 ) & (df['Insulin'+"_median"]==0), 'Insulin'+"_median"] = 39
df.loc[(df['Class Variable'] == 1 ) & (df['Insulin'+"_median"]==0), 'Insulin'+"_median"] = 0
check_dist('Insulin_median','Insulin')
#full mean
df['Insulin'+"_full_mean"]=df['Insulin'].copy().replace(0,df['Insulin'].mean())
check_dist('Insulin_full_mean','Insulin')
#full median
df['Insulin'+"_full_median"]=df['Insulin'].copy().replace(0,df['Insulin'].median())
check_dist('Insulin_full_median','Insulin')
```

```python

#equating Tricep with mean and median values with respect to thier class variables
df['Triceps'+"_mean"]=df['Triceps'].copy()
df.loc[(df['Class Variable'] == 0 ) & (df['Triceps'+"_mean"]==0), 'Triceps'+"_mean"] = 19.664000
df.loc[(df['Class Variable'] == 1 ) & (df['Triceps'+"_mean"]==0), 'Triceps'+"_mean"] = 22.164179
check_dist('Triceps_mean','Triceps')
#median
df['Triceps'+"_median"]=df['Triceps'].copy()
df.loc[(df['Class Variable'] == 0 ) & (df['Triceps'+"_median"]==0), 'Triceps'+"_median"] = 39
df.loc[(df['Class Variable'] == 1 ) & (df['Triceps'+"_median"]==0), 'Triceps'+"_median"] = 0
check_dist('Triceps_median','Triceps')
#median
df['Triceps'+"_full_median"]=df['Triceps'].replace([0],df['Triceps'].median())
check_dist('Triceps_full_median','Triceps')
#mean
df['Triceps'+"_full_mean"]=df['Triceps'].replace([0],df['Triceps'].mean())
check_dist('Triceps_full_mean','Triceps')
#random
df['Triceps'+"_random"]=df['Triceps'].copy().replace([0],np.NaN)
##It will have the random sample to fill the na
random_sample=df['Triceps'].dropna().sample(df['Triceps'].isnull().sum(),random_state=0)
##pandas need to have same index in order to merge the dataset
random_sample.index=df[df['Triceps'].isnull()].index
df.loc[df['Triceps'].isnull(),'Triceps'+'_random']=random_sample
check_dist('Triceps_random','Triceps')
```
